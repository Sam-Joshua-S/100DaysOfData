##Selecting using loc

```python
#Specifying all rows in columns of weeks and datascience
df.loc[:,['Week','DataScience']]
```

###Specifying few columns

```python
print(df.loc[[0,1,2],['Week', 'DataScience']])
```

##Using iloc 

```python
print(df.iloc[250:260,])
# Using iloc now the last element is not printed
```

##Iloc using negative indexing

```python
df.loc[-1:,:]
```

#Condition Selection

```python
df.Week == '13/09/2020'
# Identifies the row that has the value
```

```python
# Now passing that to the data frame will print that row
df.loc[df.Week == '13/09/2020']
```

```python
df.loc[(df.DataScience >= 30)]
```

##Using multiple condition

```python
df.loc[(df.DataScience >= 30) & (df.MachineLearning >= 30)]
```

#Identifying row that are null

```python
df.loc[df.DataScience.notnull()]
```

```python
df.loc[df.DataScience.isnull()]
```
