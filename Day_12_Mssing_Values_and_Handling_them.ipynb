```python
import pandas as pd
import numpy as np

df=pd.read_csv(r"https://raw.githubusercontent.com/rsharankumar/Learn_Data_Science_in_100Days/master/Day12%20-%20Missing%20Values%20and%20Handling%20Them/Data/SearchTrendData_WithMissing.csv")
df.head()
```

```python
t_data=df[:20]
t_data=t_data.reset_index(drop=True)
t_data
```

```python
print(t_data.isnull())
#print(t_data.notnull())
#print(t_data['DataScience'].isnull())
```

```python
print(t_data['DataScience'].sum())
#Missing values treated zero here
```

##Fetching all record with missing value

```python
t_data[pd.isnull(t_data.eLearning)]
```

##Fetching all rows with missing values

```python
t_data[t_data.isnull().any(axis=1)]
```

##Replacing missing values with 0

```python
print(t_data.fillna(0))
```

##Replacing missing value with mean

```python
print(t_data['eLearning'].fillna(t_data['eLearning'].mean))
```

##Use pad/forward fill to replace missing values 

```python
print(t_data.fillna(method='pad'))
```

```python
print(t_data.fillna(method='backfill'))
```

```python
 print(t_data.dropna())
```
