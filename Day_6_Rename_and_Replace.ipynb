#Topics to be covered


*   Rename - To rename index or a column
*   Replace - To replace a value with new value



```python
import pandas as pd
import numpy as np
trend_data=pd.read_csv(r"https://raw.githubusercontent.com/rsharankumar/Learn_Data_Science_in_100Days/master/Day13%20-%20Rename%20and%20Replace/Data/SearchTrendData.csv")
trend_data.head()
```

##Rename a Column of a dataframe

```python
td1=trend_data.rename(columns={"MachineLearning":'ML'})
td1.head()
```

##Rename Index

```python
td2=trend_data.rename(index={0:'a',1:'b'})
td2.head()
```

##Giving name to axis

```python
td3=trend_data.rename_axis("Dates",axis="rows").rename_axis("Topics",axis="columns")
td3.head()
```

##Renaming by  Passing the new values

```python
td4=trend_data
td4.columns=['WK','eL','DS','ML','AI','DL']
print(td4)
```

##Default string in the header

```python
td5=trend_data
td5.columns = ['$Week','eLearn','$DS','$ML','AI','DL']
td5.columns=td5.columns.str.replace('$','')
print(td5)
```

##Replace Using LOC

```python
t_data=trend_data[-10:]
t_data=t_data.reset_index(drop=True)
print(t_data)
```

```python
#Using loc
print(t_data.loc[0,'DS'])
```

```python
t_data.loc[0,'DS']=40
t_data.head()
```

##Replace Using AT

```python
#Replace using at
t_data.at[1,'DataScience'] = 40
t_data.head()
```

##Using Replace

```python
t_data.replace(40,42)
```

```python
t_data.head()
```

##Replace with inplace 

```python
t_data.replace(40,42,inplace=True)
t_data.head()
```

##Replace a list of item

```python
t_data.replace(to_replace=[42,21,15],value=[142,121,115])
```
