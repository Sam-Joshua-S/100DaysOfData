#Importing Library

```python
import gym
import os 
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
import tqdm as notebook_tqdm
```

Load Environment

```python
env_name = 'CartPole-v1'
env = gym.make(env_name)
```

```python
episodes = 5
for episode in range(1,episodes+1):
    state = env.reset()
    done = False
    score = 0 
     
    while not done:
        env.render()
        action = env.action_space.sample()
        new_state, reward, done, info = env.step(action)
        score += reward
    print('Episode:{}  Score:{}'.format(episode,score))
env.close()
```

Action space is about what are possible move are available

Num                              Action

0                              Push cart to the left

1                              Push cart to the right



```python
env.action_space
```

```python
env.step(action)
```

Observation space is about the description of state like  Cart Position   Cart Velocity Pole Angle  Pole Angular Velocity

```python
env.observation_space
```

```python
env.observation_space.sample()
```

Train an RL model

```python
log_path = os.path.join("rl","logs")
```

```python
log_path
```

```python
env = gym.make(env_name)
env = DummyVecEnv([lambda:env])
model = PPO('MlpPolicy',env,verbose=1,tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=30000)
```

Save and reload model

```python
PPO_PATH = os.path.join('rl','Saved Models','PPO_Model_CartPole')
```

```python
model.save(PPO_PATH)
```

```python
del model
```

```python
model = PPO.load(PPO_PATH,env = env)
```

Evaluation

```python
evaluate_policy(model,env,n_eval_episodes=10,render=True)
```

here average is 500 and std is 0.0

```python
env.close()
```

Test Model

```python
episodes = 5
for episode in range(1,episodes+1):
    obs = env.reset()
    done = False
    score = 0
     
    while not done:
        env.render()
        action , _ = model.predict(obs)
        obs , reward ,done , info = env.step(action)
        score += reward
    print("Epsides: {}   Score:{}".format(episode,score))
env.close()
```


Adding a Callback to the traing stage

```python
from stable_baselines3.common.callbacks import EvalCallback,StopTrainingOnRewardThreshold
```

```python
save_path = os.path.join('rl','Saved Models')
```

```python
stop_callback = StopTrainingOnRewardThreshold(reward_threshold=200,verbose=1)
eval_callback = EvalCallback(env,callback_on_new_best=stop_callback,eval_freq=10000,
                            best_model_save_path=save_path,verbose=1)
```

```python
model = PPO('MlpPolicy',env,verbose=1,tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=10000,callback=eval_callback)
```

Policies

```python
net_arch = [dict(pi=[128,128,128,128],vf=[128,128,128,128])]
```

```python
model = PPO('MlpPolicy',env,verbose=1,tensorboard_log=log_path,policy_kwargs={'net_arch':net_arch})
```

```python
model.learn(total_timesteps=10000,callback=eval_callback)
```
